```python
data = pd.read_csv("imports-85.data.csv")
print(data)
```

```python
data.head()
```

```python
brush = alt.selection_interval()
```

```python
points = alt.Chart(data).mark_point().encode(
    x='gas:O',
    y='convertible:O',
    color=alt.condition(brush, 'Origin:N', alt.value('lightgray'))
).add_selection(
    brush
)

points
```

```python
bars = alt.Chart(data).mark_bar().encode(
    y='gas:N',
    color='type:N',
    x='convertible:O'
)

bars
```
